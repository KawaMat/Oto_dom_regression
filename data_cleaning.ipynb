```python
def calc_floor(row):
    if('/' in row):
        if('> 10' in row.split('/')[0]):
            #print(row)
            try:
                return np.random.randint(11, (float(row.split('/')[1]) + 2))
            except:
                return row.split('/')[1]
        else:
            return row.split('/')[0]
    elif('> 10' in row):
        return np.random.randint(11, 16)
    else:
        return row
```

```python
# numer_pietra
df_clear['numer_pietra'] = df_clear['Piętro'].apply(lambda x: calc_floor(x))
```

```python
# liczba_pieter
def calc_floorNo(row):
    if('/' in row):
        if('> 10' in row.split('/')[1]):
            return np.random.randint(float(row.split('/')[0]), 16)
        else:
            return row.split('/')[1]
    else:
        return row
```

```python
# liczba_pieter
df_clear['liczba_pieter'] = df_clear['Piętro'].apply(lambda x: calc_floorNo(x))
```
