```python
# Load packages
import pandas as pd
import numpy as np
```

```python
df_map = pd.read_csv(r'/content/drive/MyDrive/OTO_DOM_REGRESSION/data_scraping.csv', sep = ';')
```

```python
df_map
```

```python
# Drop duplicates based on title columns
df_map_clear = df_map.drop_duplicates(subset=['Tytul'])
```

```python
df_map_clear = df_map_clear.reset_index(drop=True)
```

```python
df_map_clear
```

```python
# Create html map
import folium
from geopy.geocoders import Nominatim
```

```python
df_map_subset = df_map_clear.iloc[0:200]
```

```python
df_map_subset
```

```python
geolocator = Nominatim(user_agent='usr')
```

```python
loc_and_label = [data for data in zip(df_map_subset['Lokalizacja'], df_map_subset['Cena'], df_map_subset['Powierzchnia'], df_map_subset['Rok budowy'])]
```

```python
loc_and_label = list(zip(*loc_and_label))
```

```python
loc = loc_and_label[0]
```

```python
labels = [str('Cena: '+str(price)+'\nPowierzchnia: '+str(surface)+'\nRok budowy: '+str(year)) for price, surface, year in zip(loc_and_label[1], loc_and_label[2], loc_and_label[3])]
```

```python
labels[0]
```

```python
locations = [geolocator.geocode(location) for location in loc]
```

```python
location_coordinations = list(zip(*[(location.latitude, location.longitude) if(location!=None) else ('None','None') for location in locations]))
```

```python
# Mutalbe (list)
location_coordinations = [list(element) for element in location_coordinations]
```

```python
location_coordinations
```

```python
# Delete None values (no way to display it on map)
indices = [index for index, value in enumerate(location_coordinations[0]) if value == 'None']
```

```python
indices.sort(reverse=True)
indices
```

```python
for index_del in indices:
    del location_coordinations[0][index_del]
    del location_coordinations[1][index_del]
    del labels[index_del]
```

```python
location_map = folium.Map(location= [location_coordinations[0][0], location_coordinations[1][0]], zoom_start=12)
```

```python
for element_index in range(len(labels)):
    folium.Marker(location=[location_coordinations[0][element_index], location_coordinations[1][element_index]], 
                 popup = str('<i>'+labels[element_index]+'</i>')).add_to(location_map)
```

```python
location_map.save(r'/content/drive/MyDrive/OTO_DOM_REGRESSION/otoMap_200.html')
```
